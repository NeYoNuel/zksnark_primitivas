```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el polinomio de Lagrange
def lagrange_interpolation(x_known, y_known, x_interp):
    result = 0
    n = len(x_known)
    for i in range(n):
        term = y_known[i]
        for j in range(n):
            if i != j:
                term *= (x_interp  + (-x_known[j])%5) / (x_known[i]%5 + (-x_known[j])%5)
        result += term
    return result

if __name__ == "__main__":
    # Datos de entrada (puntos conocidos)
    x = np.array([0, -2, 2])
    y = np.array([4, 1, 3])

    # Puntos para la interpolación
    x_interp = np.linspace(-5, 5, 100)
    y_interp = lagrange_interpolation(x, y, x_interp)

    # Gráfico
    plt.plot(x_interp, y_interp, label='Interpolación de Lagrange')
    plt.scatter((x), (y), color='red', label='Puntos conocidos')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Interpolación de Lagrange')
    plt.legend()
    plt.grid()
    plt.show()
```
